```python
silver_table_name = 'crime_record_silver'

date_gold_table_name = 'dim_date_gold'
desc_gold_table_name = 'dim_desc_gold'
record_gold_table_name = 'fact_record_gold'
```

```python
silver_df = spark.read.table(silver_table_name)
```

```python
%run nb_03_dim_date_wrangler
```

```python
dim_date_gold_df = DimDateWrangler.extract_silver_df(silver_df)
DimDateWrangler.create_delta_table(spark, date_gold_table_name)

dim_date_table = DeltaTable.forName(spark, date_gold_table_name)
DimDateWrangler.upsert_delta_table(dim_date_table, dim_date_gold_df)
```

```python
%run nb_03_dim_desc_wrangler
```

```python
dim_desc_gold_df = DimDescWrangler.extract_silver_df(silver_df)
DimDescWrangler.create_delta_table(spark, desc_gold_table_name)
dim_desc_gold_df = DimDescWrangler.setup_id(spark, desc_gold_table_name, dim_desc_gold_df)

dim_desc_table = DeltaTable.forName(spark, desc_gold_table_name)
DimDescWrangler.upsert_delta_table(dim_desc_table, dim_desc_gold_df)
```

```python
%run nb_03_fact_record_wrangler
```

```python
fact_record_gold_df = FactRecordWrangler.extract_silver_df(silver_df, spark.read.table(desc_gold_table_name))
FactRecordWrangler.create_delta_table(spark, record_gold_table_name)

fact_record_table = DeltaTable.forName(spark, record_gold_table_name)
FactRecordWrangler.upsert_delta_table(fact_record_table, fact_record_gold_df)
```
