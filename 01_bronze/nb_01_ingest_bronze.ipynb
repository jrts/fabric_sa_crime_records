```python
import numpy as np
from pathlib import Path
import urllib.request
import urllib.error
import socket

base_url = 'https://data.sa.gov.au/data/dataset/860126f7-eeb5-4fbc-be44-069aa0467d11/resource'
data_url_path = '/lakehouse/default/Files/data_source/sa_crime_data_urls.csv'
output_base_url = '/lakehouse/default/Files/sa_crime_datasets'
```

```python
def url_retrieve(url: str, outfile: Path, overwrite: bool = False):
    outfile = Path(outfile).expanduser().resolve()
    if outfile.is_dir():
        raise ValueError("Please specify full filepath, including filename")
    # need .resolve() in case intermediate relative dir doesn't exist
    if overwrite or not outfile.is_file():
        outfile.parent.mkdir(parents=True, exist_ok=True)
        try:
            urllib.request.urlretrieve(url, str(outfile))
        except (socket.gaierror, urllib.error.URLError) as err:
            raise ConnectionError(
                "could not download {} due to {}".format(url, err)
            )
```

```python
data_urls = np.loadtxt(data_url_path, dtype=np.str_, skiprows=1)
```

```python
for url in data_urls:
    ds_path = f'{base_url}/{url}'
    output_path = f"{output_base_url}/{url.split('/')[-1]}"
    url_retrieve(ds_path, output_path)
```
