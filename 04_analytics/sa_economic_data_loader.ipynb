## Crash Rate Target (%, by month)

```python
crt_df = pd.read_excel(
    io=f'{base_url}/a02hist.xlsx',
    sheet_name='Data',
    header=1,
    skiprows=range(2, 14),
)

crt_df = crt_df.rename({crt_df.columns[0]:'Date'}, axis=1)
crt_df['month_year'] = crt_df['Date'].dt.to_period('M')
crt_df = crt_df.iloc[:, 2:]
crt_df = crt_df.rename({crt_df.columns[0]:'CRT'}, axis=1)
```

```python
# CRT reduced twice at March 2020, keep the last one
crt_df = crt_df.drop([72])
```

```python
crt_df = crt_df.set_index('month_year')
crt_df = crt_df.sort_index()
crt_df = crt_df.reindex(pd.period_range(crt_df.index[0], crt_df.index[-1], freq='M'))
```

```python
# fill na with last valid observation
crt_df = crt_df.ffill()
crt_df['Change'] = crt_df['CRT'].diff()
crt_df = crt_df.fillna(0.0)
# crt_df
```

## SA Gross Regional Product (GRP, $million, by year)

```python
grp_df = pd.read_excel(
    io=f'{base_url}/GRP.xlsx',
    usecols=[0, 4, 5, 6],
    header=2,
)
grp_df = grp_df.drop([24]).iloc[::-1]
```

```python
grp_columns = {
    grp_df.columns[0]: 'Year',
    grp_df.columns[1]: 'GRP',
    grp_df.columns[2]: 'Change',
    grp_df.columns[3]: 'Cumulative change'
}
grp_df = grp_df.rename(grp_columns, axis=1)
grp_df.at[23, 'Change'] = 0.0
grp_df = grp_df.astype({'Change': 'float'})
grp_df = grp_df.set_index('Year')
# grp_df
```
